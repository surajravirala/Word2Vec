```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
model=KeyedVectors.load_word2vec_format("/content/drive/MyDrive/GoogleNews-vectors-negative300.bin",binary=True,limit=500000)
```

```python
model["Names"] # Access the raw vector (mathematical representation) of a word
```

```python
model.most_similar("Dog")# Find words with the highest cosine similarity to "Dog"
```

```python
model.most_similar("man")# Find words most similar to "man"
```

```python
model.similarity("man","woman")# Calculate the cosine similarity score (0 to 1) between two words
```

```python
model.similarity("Dog","Cat")# Calculate the cosine similarity score (0 to 1) between two words
```

```python
model.similarity("Pen","Dog")
```

```python
model.doesnt_match(["sketch","pen","Lion"])# Identify the 'outlier' in a list based on semantic distance
# In this case, "Lion" (animal) vs "sketch" and "pen" (stationery)
```

```python
x=model["king"]-model["man"]+model["women"]# Famous Word2Vec Analogy: King - Man + Woman = Queen
                                            # We use positive for terms to add and negative for terms to subtract
model.most_similar([x])
```
